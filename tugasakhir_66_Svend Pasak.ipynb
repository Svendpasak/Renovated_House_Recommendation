```python
# Menjadikan nilai kolom price dari eksponensial menjadi 2 digit setelah koma
df['price'] = df['price'].map(lambda x: f"{x:.2f}").astype('float')
```

## EXPLORATORY DATA ANALYSIS (EDA)

```python
df.dtypes
```

```python
filtered_df= df[(df.yr_built>=1990)|((df.yr_built<=1990) & (df.yr_renovated>=2000))]
filtered_df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=filtered_df.dropna(subset=['yr_renovated']),
    x='yr_built',
    y='yr_renovated',
    alpha=0.5
)
plt.title('Tahun Dibangun vs Tahun Renovasi (Setelah Difilter)')
plt.xlabel('Tahun Dibangun')
plt.ylabel('Tahun Renovasi')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot Histogram Tahun Renovasi Rumah
plt.figure(figsize=(12, 6))
sns.histplot(filtered_df['yr_renovated'].dropna(), bins=30, color='salmon')
plt.title('Distribusi Tahun Renovasi Rumah (Setelah Difilter)')
plt.xlabel('Tahun Renovasi')
plt.ylabel('Jumlah Rumah')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## MENGHITUNG SKOR DAN MEMILIH 10 RUMAH TERBAIK

```python
# Menambahkan kolom baru dengan nama "price per sqft"
filtered_df['price_per_sqft']= filtered_df['price'] / filtered_df['sqft_living']
filtered_df.head()
```

```python
# Normalisasi data (semakin rendah price_per_sqft, semakin baik)
# Kita cari rumah dengan kondisi bagus & harga murah per sqft
recommendation_df = filtered_df.copy()
```

```python
# Menetapkan Skor sederhana
# Semakin tinggi condition (1-5) dan semakin rendah price_per_sqft, semakin bagus
# Tujuannya agar rumah dengan kondisi bagus & harga terjangkau mendapat skor tinggi
recommendation_df['score'] = (
    recommendation_df['condition'] * 2 -
    recommendation_df['price_per_sqft'] / 100
    )
```

```python
# Ambil 10 rumah dengan skor tertinggi
top_10 = recommendation_df.sort_values(by='score', ascending=False).head(10)
```
